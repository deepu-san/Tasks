```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
path="https://raw.githubusercontent.com/deepu-san/Tasks/main/World%20Population/world_population.csv"
df = pd.read_csv(path)
df.head(3)
```

```python
df.shape
```

```python
df['Country'].nunique()
```

```python
plt.figure(figsize = (20,8))
plt.title("Top 20 Countries based on World Population Percentage")
sns.barplot(x='Country', y="World Population Percentage", data=df.nlargest(20,"World Population Percentage"))

```

```python
plt.figure(figsize = (18,8))
plt.title("Top 20 Countries based on Growth Rate")
sns.barplot(x='Country', y="Growth Rate", data=df.nlargest(20,"Growth Rate"))

```

```python
plt.figure(figsize = (20,8))
plt.title("Top 20 Countries based on Density")
sns.barplot(x='Country', y="Density (per km²)", data=df.nlargest(20,"Density (per km²)"))

```
